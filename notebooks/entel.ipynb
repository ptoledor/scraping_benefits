```python
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
import time
```

```python
def parse_entel(url, desc_url):

    # create a new Chrome browser instance
    driver = webdriver.ChromiumEdge()

    # navigate to the website you want to scrape
    driver.get(url)

    # wait for the JavaScript to generate the content
    time.sleep(3)
    driver.implicitly_wait(15)

    # parse the HTML content of the page
    soup = BeautifulSoup(driver.page_source, 'html.parser')

    # close the browser instance
    driver.quit()

    # find all elements with the specified class
    items = soup.find_all('div', class_='col-lg-4 col-md-6 card-categorias p-3')

    df = []

    for item in items:
        id = ' '.join((str(item.find('p', class_='title-box mt-4').text).replace('\n', '')).split())
        desc1 = item.find('p', class_='txt-box mt-4 mt-md-0').text
        desc2 = item.find('a')['data-descripcion']
        id2 = item.find('a')['data-titulo']
        data_variante = item.find('a')['data-variante']
        link = 'https://www.entel.cl/beneficios/descuentos/' + data_variante + item.find('a')['href'][1:]

        df.append([id, id2, link,desc1, desc2])

    df = pd.DataFrame(df, columns=['ID', 'ID2', 'LINK', 'DESCRICION_1', 'DESCRIPCION_2'])
    df.insert(loc=0, column='ENTIDAD', value='ENTEL')
    df.insert(loc=1, column='DESCRIPCION_URL', value=desc_url)

    return df
```

```python
df1 = parse_entel('https://www.entel.cl/beneficios/descuentos/comida/', 'COMIDA')
df2 = parse_entel('https://www.entel.cl/beneficios/descuentos/deportes/', 'DEPORTES')
df3 = parse_entel('https://www.entel.cl/beneficios/descuentos/bienestar/', 'BIENESTAR')
df4 = parse_entel('https://www.entel.cl/beneficios/descuentos/tienda/', 'TIENDA')
df5 = parse_entel('https://www.entel.cl/beneficios/descuentos/entretencion/', 'ENTRETENCION')
df6 = parse_entel('https://www.entel.cl/beneficios/descuentos/sostenibilidad/', 'SOSTENIBILIDAD')
df7 = parse_entel('https://www.entel.cl/beneficios/descuentos/transporte/', 'TRANSPORTE')
df8 = parse_entel('https://www.entel.cl/beneficios/descuentos/gaming/', 'GAMING')


df_final = pd.concat([df1, df2, df3, df4, df5, df6, df7, df8], ignore_index=True)
df_final
```

```python
df_final.to_csv('entel.csv', sep=';', decimal=',', index=False, encoding='utf-8-sig')
```
